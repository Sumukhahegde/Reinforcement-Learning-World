# Reinforcement Learning

## Machine Learning

Machine learning is an application of artificial intelligence (AI) that provides systems the ability to automatically learn and improve from experience without being explicitly programmed. Machine learning focuses on the development of computer programs that can access data and use it learn for themselves.

Machine Learning is classified into three types:

### Supervised Learning:
Consider yourself as a student sitting in a classroom wherein your teacher is supervising you, how you can solve the problem or whether you are doing correctly or not. Likewise, in Supervised Learning input is provided as a labelled dataset, a model can learn from it to provide the result of the problem easily.

### Unsupervised Learning:
This learning algorithm is completely opposite to Supervised Learning. In short, there is no complete and clean labelled dataset in unsupervised learning. Unsupervised learning is self-organized learning. Its main aim is to explore the underlying patterns and predicts the output.  Here we basically provide the machine with data and ask to look for hidden features and cluster the data in a way that makes sense.

### Reinforcement Learning:
In Reinforcement Learning Problem an agent tries to manipulate the environment. The agent travels from one state to another. The agent gets the reward(appreciation) on success but will not receive any reward or appreciation on failure. In this way, the agent learns from the environment.


![Description of the three types](https://gblobscdn.gitbook.com/assets%2F-LvBP1svpACTB1R1x_U4%2F-LwQbcbsD6nnVTHSgxV1%2F-LwQcH6NWMz6kjAEqpge%2Fimage.png?alt=media&token=2ccddf94-0321-472e-a2dd-49e7ab63914d)

## Reinforcement Learning and its components:
Reinforcement learning is learning what to do; how to map situations to actions so
as to maximize a numerical reward signal. The learner is not told which actions to
take, but instead must discover which actions yield the most reward by trying them
Below we introduce a few components of a Reinforcement Learning (RL) based architecture:

Let's start of by an example to make things simple:

* Consider a Grid World like the one below with 9 cells. This will be our *Environment*

* Now consider a yellow ball placed in the top-left cell of the grid. This will be our *Agent*.

#### Environment: 
An environment is everything which isnt the Agent. Everything the Agent can interact with, either directly or indirectly is part of the environment.

#### Agent:
An Agent is the learning and acting part of a Reinforcement Learning problem, which tries to interatcs with a given Environment. Putting it simply, the 
Agent is the model which you try to design.

Now, let's continue with our example.
Consider that in the given setting each cell inside the grid is a green patch where the ball can go to by moving one cell at a time. Then all these patches where the ball can be are called the *States*.

Actions are the Agents methods which allow it to interact and change its environment, and thus transfer between states.



#### State:
Every scenario the Agent encounters in the Environment is formally called a state. The Agent transitions between different states by performing actions.

The next important concept we want to introduce is the concept of *Rewards*.
An agent works with the aim of maximizing its rewards and we are supposed to design the reward system based on what we want the agent to do.
Here we want the agent to go to the Flag patch so we will arrange the rewards in such a manner that the agent is to motivated to take the actions that take it to the Flag patch as quickly as possible.

We'll consider that the action of going over the mountain is expensive and takes more effort than going through other patches. So we set up a higher penalty for the agent when it takes an action to go over the mountain patches.


#### Reward:
A numerical value received by the Agent from the Environment as a direct response to the Agents actions. The Agents goal is to maximize the overall reward it receives during an episode, and so rewards are the motivation the Agent needs in order to act in a desired behavior.


<br>

#### Episode:
All states that come in between an initial-state and a terminal-state; for example: one game of Chess. The Agents goal it to maximize the total reward it receives during an episode. In situations where there is no terminal-state, we consider an infinite episode.
It is also worth mentioning the terminal states, which mark the end of an episode. 

<br>

#### Policy:
The policy, denoted as  (or sometimes (a|s)), is a mapping from some state s to the probabilities of selecting each possible action given that state. For example, a greedy policy outputs for every state the action with the highest expected Q-Value.

<br>

#### Value Function:
Usually denoted as V(s) (sometimes with a  subscript), the Value function is a measure of the overall expected reward assuming the Agent is in state s and then continues playing until the end of the episode following some policy .

<br>

#### Model:
Assume youre trying to learn how to play Blackjack. One approach to do that is you calculate in advance, before the game begins, the winning probabilities of all states and all the state-transition probabilities given all the possible actions, and then simply act according to you calculations. Here youre basically modeling your environment.

Next we will introduce a few more terminologies and notations:

### Expected Cumulative Reward:

So suppose you have an environment and you have an agent that interacts with the environment. The way this interaction happens, is that the Environment presents a State to the Agent. The Agent then performs an action accordingly in the environment. And then the Environment presents the Agent with a Reward value accordingly.

<img src ="State_Action_Reward.PNG" alt="ENvironment" width="350" align="left"/>



Along with this we have another goal built into the agent's framework, which is to *maximize Expected Cumulative Reward*.

A few notations to consider here are as follows:

$S_t$ = State at timestep t

$A_t$ = Action at timestep t

$R_t$ = Reward at timestep t

So suppose for the first timestep ($t = 0$), the environment and agent interact with each other and we get a set of State, Action and Reward as ($S_0, A_0, R_0$). Similarly at the other timesteps we will be getting the respective pairs ($S_1, A_1, R_1, S_2, A_2, R_2, ....... S_t, A_t, R_t$).

Now let's elaborate a little on what we mean by maximizing expected cumulative reward.
Suppose the Agent interacts with the environment through a few timesteps ($t=[0, 1, 2]$) and the rewards it received for the actions ($A_0, A_1, A_2$) is respectively ($R_0=-1, R_1=-2, R_2=2$).
Then the cumulative reward upto this point is $R_0 + R_1 + R_2 = -1$.

At this point the agent cannot do anything about its actions in the past to change the rewards it has received. What it can do is change its action is such a way that it can maximize the cumulative reward it can receive in the future ($R_3 + R_4 +..... + R_t$) i.e. Expected Cumulative Reward ($G_t$). We say "expected" because the agent cannot always predict what reward it will receive and so it takes an estimate of what rewards it can expect.

Here in the above example we have,

$G_2 = R_3 + R_4 + .... + R_t$

### Discounted Return

This concept might be a little bit tricky to follow on with so try reading this part below a couple of times till you feel confident about it.

Consider a puppy that will be our agent and we interact with the puppy as part of the environment. 

<img src ="Puppy1.PNG" alt="ENvironment" width="250" align="left"/>




We as part of the environment present the puppy with a choice as the *state*. A choice between taking a Marshmallow right now or *maybe* get a marshmallow tomorrow.

<img src ="Puppy2.PNG" alt="ENvironment" width="250" align="left"/>



For the puppy it would make more sense to take a Marshmallow right now since that option guarantees more reward rather than a possibility of some reward in the future. 


Now if we offer the puppy a different choice between taking a Marshmallow right now or getting one for the next two days, the puppy would might want to still take the Marshmallow right now as thinking that far ahead in the future does not seem reasonable when it can have a Marshmallow this instant.
<img src ="Puppy3.PNG" alt="Puppy2" width="250" align="left"/>





So in a sense the idea we are trying to get to here is that although the agent would like to maximize epected cumulative reward, it makes sense to be more concerned about the rewards closer in the future than the distant ones.

So we update our Expected Cumulative Reward formula from:

$G_t = R_{t+1} + R_{t+2} + R_{t+3} ...$

to

$G_t = R_{t+1} + \gamma*R_{t+2} + \gamma^2*R_{t+3} ...$

Here $G_t$ is called the discounted return and $\gamma$ is the factor by which we reduce the weightage of the rewards in the future.

### Markov Decision Process

In a finite MDP, the sets of states, actions, and rewards ($\mathcal{S}, \mathcal{A}, \mathcal{R}$) all have a finite
number of elements. In this case, the random variables $R_t$ and $S_t$ have well defined discrete probability distributions dependent only on the preceding state and action. That is, for particular values of these random variables, $s' \in \mathcal{S}$ and $r' \in \mathcal{R}$, there is a probability
of those values occurring at time $t$, given particular values of the preceding state and action:
$p(s',r|s,a) \doteq \mathbb{P}(S_{t+1}=s', R_{t+1}=r|S_{t} = s, A_{t}=a)$
for all $s'
, s \in \mathcal{S}, r \in \mathcal{R}$, and $a \in \mathcal{A}(s)$. The function p defines the dynamics of the MDP.

A (finite) Markov Decision Process (MDP) is defined by:

* a (finite) set of states $\mathcal{S}$ (or $\mathcal{S}^+$ , in the case of an episodic task)
* a (finite) set of actions $\mathcal{A}$
* a set of rewards $\mathcal{R}$
* the one-step dynamics of the environment
* the discount rate $\gamma \in [0,1]$

### Policy

This is an important concept in RL, one that you will work towards optimizing.
To define, you can consider a Policy as a mapping between a set of States and a set of Actions

$ \pi : \mathcal{S} \rightarrow \mathcal{A}$

If that was a bit technical, we can discuss about Policy on a higher level.
You can consider Policy as a Factory that takes a State as an input and returns a corresponding Action that the agent will follow.


<img src ="Policy_Factory.PNG" alt="Policy_Factory" width="550" align="left"/>




If an Agent needs to keep track of its Strategy, all it needs to do is build this factory, or specify its *mapping*.

There are two types of Policy:

1. Deterministic Policy: The policy accepts a state and returns an action.

   $ \pi : \mathcal{s} \rightarrow \mathcal{A} $

    <img src ="Det_Policy.PNG" alt="Det_Policy" width="550" align="left"/>



2. Stochastic Policy: Here the policy accepts a State and an Action as input and returns the probability that the agent takes action a while in state s

   $ \pi(s|a) = \mathbb{P}(A_t = a|S_t = s) $

   <img src ="Stoc_Policy.PNG" alt="Stoc_Policy" width="550" align="left"/>


So a Policy is basically the solution that the Agent will follow whenever it enters a new state with the aim of maximizing expected cumulative reward.

### State Value Function

Now that we have an understanding of how a policy works. Lets go on to understand how we can determine the best policy.

Let's go back to our Grid World example that we introduced earlier.

We had an agent placed at the top left corner of a grid world environment with multiple cells inside the grid being the states and the possible sets of moves that the agent can make being the actions.

<img src ="Reward.PNG" alt="Reward" width="550" align="left"/>
![Reward]()


Now lets consider a bad policy so that once we understand it, we can work towards improving it. Consider the policy being that the agent starts from the top left patch and goes over all the patches in a roundabout manner till it reaches the goal and it collects the rewards along the way.
<img src ="BadPol1.PNG" alt="BadPol1" width="350" align="left"/>




   Consider the discount rate to be $\gamma = 1$

   Then the total reward we have is 

   (-1) + (-1) + (-1) + (-3) + (-1) + (-1) + (-3) + 5 = -6



Then we can consider this as the reward the agent will receive if it follow the given policy from the 1st cell.
<img src ="BadPol2.PNG" alt="BadPol2" width="350" align="left"/>


Now lets consider if the agent starts from the second cell at the top and continues with the same policy. 
<img src ="BadPol3.PNG" alt="BadPol3" width="350" align="left"/>




Here the total reward we have is 

(-1) + (-1) + (-3) + (-1) + (-1) + (-3) + 5 = -5

We can keep track of this value as well signifying the reward the agent will receive if it starts from this position and follows the policy to the goal.
<img src ="BadPol4.PNG" alt="BadPol4" width="350" align="left"/>



Similarly we can calculate this value for all cells and get the following values:
<img src ="BadPol5.PNG" alt="BadPol5" width="350" align="left"/>


You can think of this grid of numbers as function of the environment state

<img src ="BadPol6.PNG" alt="BadPol6" width="350" align="left"/>


For each state you have a corresponding number. And we call this function as the State Value Function.

The state-value function for a policy $\pi$ is denoted $v_\pi$. For each state $s \in \mathcal{S}$, it yields the expected return if the agent starts in state $s$ and then uses the policy to choose its actions for all time steps. 

$v_\pi(s) \doteq \text{} \mathbb{E}\pi[G_t|S_t=s]$

### Bellman Equation

In the above example, at any state we have the immediate reward, the value of the state and the immediate reward that follows.
<img src ="Bellman1.PNG" alt="Bellman1" width="450" align="left"/>





Bellman Equation states that the value of the start state must equal the
(discounted) value of the expected next state, plus the reward expected along the way.

<img src ="Bellman2.PNG" alt="Bellman2" width="550" align="left"/>


$v_\pi(s) = \text{} \mathbb{E}\pi[R{t+1} + \gamma v_\pi(S_{t+1})|S_t =s]$

### Optimal Solution

An optimal solution is the solution you want your agent to aim for. In our example it is the route that the agent can take to reach to the goal position as fast as possible and by maximizing its rewards. 
* A policy $\pi$ is defined to be better than or equal to a policy $\pi$ if and only if $v_{\pi}(s) \geq v_\pi(s)v$ for all $s\in\mathcal{S}$.
* An optimal policy $\pi_*$ satisfies $\pi_* \geq \pi$ for all $\pi$. An optimal policy is guaranteed to exist but may not be unique.
* All optimal policies have the same state-value function $v_*$, called the optimal state-value function.

### Action Value Function
* The action-value function for a policy $\pi$ is denoted $q_\pi$. For each state $s \in\mathcal{S}$ and action $a \in\mathcal{A}$, it yields the expected return if the agent starts in state $s$, takes action $a$, and then follows the policy for all future time steps.

### 1. Start the Environment


```python
from unityagents import UnityEnvironment
import numpy as np
import time
from collections import deque
import matplotlib.pyplot as plt
import torch

#env = UnityEnvironment(file_name="/data/Banana_Linux_NoVis/Banana.x86_64") #to run on udacity workspace
env = UnityEnvironment(file_name="Banana_Windows_x86_64/Banana.exe") #to run locally
```

Environments contain **_brains_** which are responsible for deciding the actions of their associated agents. Here we check for the first brain available, and set it as the default brain we will be controlling from Python.

```python
# get the default brain
brain_name = env.brain_names[0]
brain = env.brains[brain_name]
```

### 2. Examine the State and Action Spaces

Run the code cell below to print some information about the environment.

```python
# reset the environment
env_info = env.reset(train_mode=True)[brain_name]

# number of agents in the environment
print('Number of agents:', len(env_info.agents))

# number of actions
action_size = brain.vector_action_space_size
print('Number of actions:', action_size)

# examine the state space 
state = env_info.vector_observations[0]
print('States look like:', state)
state_size = len(state)
print('States have length:', state_size)
```

### 3. Take Random Actions in the Environment

In the next code cell, we will learn how to use the Python API to control the agent and receive feedback from the environment.

Note that **in this coding environment, we will not be able to watch the agent while it is training**, and we should set `train_mode=True` to restart the environment.

```python
env_info = env.reset(train_mode=False)[brain_name] # reset the environment
state = env_info.vector_observations[0]            # get the current state
score = 0                                          # initialize the score
while True:
    action = np.random.randint(action_size)        # select an action
    env_info = env.step(action)[brain_name]        # send the action to the environment
    next_state = env_info.vector_observations[0]   # get the next state
    reward = env_info.rewards[0]                   # get the reward
    done = env_info.local_done[0]                  # see if episode has finished
    score += reward                                # update the score
    state = next_state                             # roll over the state to next time step
    if done:                                       # exit loop if episode finished
        break
    
print("Score: {}".format(score))
```

## Training the model

```python
#Training the model
from dqn_agent import Agent

agent = Agent(state_size=37, action_size=4, seed=0)
#print([p for p in agent.qnetwork_local.parameters()])

# #debug
# state = env.reset()
# print(agent.qnetwork_local(state))
# print(agent.qnetwork_local.forward(state))
# #end debug

filename_to_save = 'ddqn_checkpoint.pth'
final_eps = 0.01
def dqn(n_episodes=2000, max_t=1000, eps_start=1.0, eps_end=final_eps, eps_decay=0.995):
    """Deep Q-Learning.
    
    Params
    ======
        n_episodes (int): maximum number of training episodes
        max_t (int): maximum number of timesteps per episode
        eps_start (float): starting value of epsilon, for epsilon-greedy action selection
        eps_end (float): minimum value of epsilon
        eps_decay (float): multiplicative factor (per episode) for decreasing epsilon
    """
    scores = []                        # list containing scores from each episode
    scores_window = deque(maxlen=100)  # last 100 scores
    eps = eps_start                    # initialize epsilon
    for i_episode in range(1, n_episodes+1):     
        env_info = env.reset(train_mode=True)[brain_name] # reset the environment
        state = env_info.vector_observations[0]            # get the current state
        score = 0   # initialize the score
        for t in range(max_t): #this could also be while True instead
            action = agent.act(state, eps)        # select an action
            env_info = env.step(action)[brain_name]        # send the action to the environment
            next_state = env_info.vector_observations[0]   # get the next state
            reward = env_info.rewards[0]                   # get the reward
            done = env_info.local_done[0]                  # see if episode has finished
            agent.step(state, action, reward, next_state, done) #do the learning

            score += reward                                # update the score
            state = next_state                             # roll over the state to next time step
            if done:                                       # exit loop if episode finished
                break
        scores_window.append(score)       # save most recent score
        scores.append(score)              # save most recent score
        eps = max(eps_end, eps_decay*eps) # decrease epsilon
        print('\rEpisode {}\tAverage Score: {:.2f}'.format(i_episode, np.mean(scores_window)), end="")
        if i_episode % 100 == 0:
            print('\rEpisode {}\tAverage Score: {:.2f}'.format(i_episode, np.mean(scores_window)))
        if np.mean(scores_window)>=25.0:
            print('\nEnvironment solved in {:d} episodes!\tAverage Score: {:.2f}'.format(i_episode-100, np.mean(scores_window)))
            torch.save(agent.qnetwork_local.state_dict(), filename_to_save)
            break
    torch.save(agent.qnetwork_local.state_dict(), filename_to_save) #for debug only
    return scores

strt = time.time()
scores = dqn()
print('Training Time is {}'.format(time.time()-strt))

# plot the scores
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(np.arange(len(scores)), scores)
plt.ylabel('Score')
plt.xlabel('Episode #')
plt.show()
```

## Testing the model

```python

from dqn_agent import Agent
filename_to_load = './ddqn_checkpoint.pth'
final_eps = 0.01

agent = Agent(state_size=37, action_size=4, seed=0)
# load the weights from file
#agent.qnetwork_local.load_state_dict(torch.load(filename_to_load))

#Since the model is trained on gpu, need to load all gpu tensors to cpu:
agent.qnetwork_local.load_state_dict(torch.load(filename_to_load, map_location=lambda storage, loc: storage))


#print([p for p in agent.qnetwork_local.parameters()])

# #debug
# state = env.reset()
# print(agent.qnetwork_local(state))
# print(agent.qnetwork_local.forward(state))
# #end debug

num_episodes = 3
scores = []
for i_episode in range(1,num_episodes+1):
    env_info = env.reset(train_mode=False)[brain_name] # reset the environment
    state = env_info.vector_observations[0]            # get the current state
    score = 0                                          # initialize the score
    while True:
        action = agent.act(state, eps=final_eps)        # select an action
        env_info = env.step(action)[brain_name]        # send the action to the environment
        next_state = env_info.vector_observations[0]   # get the next state
        reward = env_info.rewards[0]                   # get the reward
        done = env_info.local_done[0]                  # see if episode has finished
        #agent.step(state, action, reward, next_state, done) #do the learning

        score += reward                                # update the score
        state = next_state                             # roll over the state to next time step
        if done:                                       # exit loop if episode finished
            print('Episode {}: {}'. format(i_episode, score))
            scores.append(score)
            break

print('All the scores{}'.format(scores))
print("Mean Score: {}".format(np.mean(scores)))
```
